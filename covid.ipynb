## COVID-19 Impact on Pakistan – Data Analysis Project

This project analyzes the effect of COVID-19 in Pakistan using a real-world dataset.  
We'll examine trends, visualize insights, and understand how the pandemic unfolded using Python, Pandas, Seaborn, and Matplotlib.


IMPORT Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

loading the dataset

```python
df = pd.read_csv('covid_cleaned.csv')
df
```

- Filtered only Pakistan from global dataset

- Converted date to datetime format  
- Handled null values by filling with 0  

visualizing total cases overtime 

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df , x='Survey Date',y='total_cases',color='green')
plt.title('Total cases over date')
plt.xlabel('date')
plt.ylabel('cases')
plt.show()
```

total deaths over time 

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df , x='Survey Date',y='total_deaths',color='red')
plt.title('Total deaths in pakistan')
plt.xlabel('Date')
plt.ylabel('Deaths')
plt.show()
```

new cases and new deaths 

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df , x='Survey Date', y='new_cases', label='New Cases')
sns.lineplot(data=df , x='Survey Date' , y='total_deaths', label='New deaths', color='orange')
plt.legend()
plt.title('new cases over new deaths')
plt.show()
```

peak days and peak cases 

```python
print("peak new caes")
print(df[df['new_cases'] == df["new_cases"].max()])
print("peak deaths")
print(df[df["total_deaths"]==df["total_deaths"].max()])
```

total cases over total deaths overview

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df,x='Survey Date', y='total_cases',label='Total Caes')
sns.lineplot(data=df,x='Survey Date',y='total_deaths',label='Total Deaths',color='black')
plt.title('Total cases over deaths')
plt.legend()
plt.show()
```

heat map

```python
plt.figure(figsize=(10,4))
sns.heatmap(df.isnull(), cbar=False, cmap="viridis")
plt.title("Null Value Heatmap")
plt.show()
```
